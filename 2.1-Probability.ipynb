১.১ সাধারন প্রবাবিলিটি নির্ণয়ের ফাংশন। 

```python
def P(favourable_outcome, total_outcome):
  p=favourable_outcome/total_outcome
  print("Probability is",round(p, 3))
```

১.২ লুডু খেলায় ৬ উঠার প্রবাবিলিটি কত ?

```python
favourable_outcome=1
total_outcome=2
P(favourable_outcome, total_outcome)
```

১.৩ ল অব লার্জ নাম্বার : কয়েন টস এক্সপেরিমেন্ট

```python
from random import randint 
trails= 100
outcomes=[]
for x in range (trails):
  outcomes.append(randint(0,1))

head=outcomes.count(0)
tail=outcomes.count(1)

print ("Number of Trails= ", trails)
print ("Head=", head)
print ("Tail=", tail)
```

২.১ প্রবাবিলিটির গুনের সূত্রঃ ইন্ডিপেন্ডেন্ট ইভেন্টের ক্ষেত্রে

```python
def a_and_b(a,b,total):
  pa=a/total
  pb=b/total
  ab=pa*pb
  print("P(A and B)=",round(ab, 3))
```

২.২ একটি পাত্রে ৩ টি নীল মার্বেল , ২ টি লাল এবং ৪ টি হলুদ মার্বেল রাখা আছে। এখন আপনাকে বলা হোল প্রথমে একটি মার্বেল তুলবেন , সেটির রং লিখে রাখবেন এবং পাত্রে ফেরত রাখবেন। এরপর আবার আরেকটি মার্বেল তুলবেন এবং সেটির রংও লিখে রাখবেন। এখন প্রথম মার্বেলটি লাল হবার পরে দ্বিতীয়টি নীল হবার প্রবাবিলিটি কত ?

```python
red=2
blue=3
total=9
a_and_b(red,blue,total)
```

২.৩ প্রবাবিলিটির গুনের সূত্রঃ ডিপেন্ডেন্ট  ইভেন্টের ক্ষেত্রে

```python
def a_and_b_conditional(a,b,total):
  pa=a/total
  remaining_total=total-1
  remaining_a_probability=b/remaining_total
  ab=pa*remaining_a_probability
  print("P(A and B)=",round(ab, 3))
```

২.৪ একটি বাক্সে ৫ টি সবুজ পেন্সিল এবং ৭ টি হলুদ পেন্সিল আছে। ধরুন আপনি প্রথমে একটি হলুদ পেন্সিল তুললেন এবং সেটি আর ফেরত রাখলেন না । এবার দ্বিতীয় পেন্সিল তুললে সেটিও হলুদ হবার প্রবাবিলিটি কত ? 

```python
yellow=7
remaining_yellow=6
total=12
a_and_b_conditional(yellow,remaining_yellow,total)
```

৩.১ প্রবাবিলিটির যোগের সূত্রঃ মিউচুয়াল এক্সক্লুসিভ ইভেন্টের ক্ষেত্রে

```python
def a_or_b(a,b,total):
  pa=a/total
  pb=b/total
  ab=pa+pb  
  print("P(A or B)=",round(ab, 3))
```

৩.২ মনে করুন আপনাকে ৫২ টি তাসের একটি প্যাকেট থেকে যেকোনো একটি তাস বাছাই করতে বলা হোল। আপনার বাছাইকৃত তাসটি একটি কিং অথবা কুইন হবে তার প্রবাবিলিটি কত ?

```python
king=4
queen=4
total=52
a_or_b(king,queen,total)
```

৩.৩ প্রবাবিলিটির যোগের সূত্রঃ কমন ইভেন্টের ক্ষেত্রে

```python
def a_or_b_common(a,b,anb,total):
  pa=a/total
  pb=b/total
  a_and_b=anb/total
  ab=pa+pb-a_and_b
  print("P(A or B)=",round(ab, 3))
```

৩.৪ কোন একটি ক্লাসে মোট ৩০ জন শিক্ষার্থী আছে, যাদের ভেতরে ১৭ জন ছেলে এবং ১৩ জন মেয়ে। ক্লাস টেস্টে ৪ জন ছেলে এবং ৫ জন মেয়ে A+ পেয়েছে। এখন যদি আপনাকে র‍্যান্ডমলি একজন শিক্ষার্থী বাছাই করতে বলা হয় তাহলে ঐ শিক্ষার্থীর মেয়ে অথবা A+ পাবার প্রবাবিলিটি কত হবে ?

```python
Aplus=9
girl=13
girl_and_Aplus=5
total=30
a_or_b_common(Aplus,girl,girl_and_Aplus,total)
```

৪। বেইজ থিওরাম

```python
def bayes(a,b,ba):
  ab= a*ba/b
  print("P(A | B)=",round(ab, 3))
```
